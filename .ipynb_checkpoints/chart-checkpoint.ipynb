```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_json('nba.save.json')
```

```python
df.sample()
```

```python
dfs = pd.read_json('salaries.save.json')
```

```python
dfs.sample()
```

```python
dfsr = dfs.rename(columns={'player': 'PLAYER', 'team': 'TEAM'})
```

```python
rookiesDf = pd.read_json("rookies.json")
# rookiesDf["PLAYER"] = rookiesDf["PLAYER"].astype("|S")
```

```python
# df["PLAYER"] = df["PLAYER"].astype("|S")
# dfsr["PLAYER"] = dfsr["PLAYER"].astype("|S")
```

```python
dfsr.head()
```

```python
dfTradional = pd.read_json('nba.tradional.json')
dfTradional.drop(['TEAM','AGE','GP','W','L','MIN'], inplace=True, axis=1)
```

```python
dfAllStar = pd.read_json('wikipedia.json')
```

```python
dfb = df.merge(dfsr,on='PLAYER',how="left")
dfb = dfb.merge(dfTradional,on='PLAYER',how="left")
dfb = dfb.merge(dfAllStar,on='PLAYER',how="left")
dfb = dfb[~dfb.PLAYER.isin(rookiesDf.PLAYER)]
```

```python
len(dfb)
```

```python
dfb.ALLSTAR = dfb.ALLSTAR.fillna(0)
```

```python
dfb.head()
```

```python
dfbd = dfb.dropna()
```

```python
len(dfbd)
```

```python
dfbd.head()
```

```python
dfbdf = dfbd[dfbd.TEAM_x == dfbd.TEAM_y]
```

```python
len(dfbdf)
```

```python
dfbdf.columns
```

```python
dfp = dfbdf.drop(['TEAM_y'], inplace=False, axis=1)
```

```python
dfp = dfp.rename(columns={'TEAM_x': 'TEAM'})
```

```python
dfp.head()
```

```python
dfp.columns
```

```python
dfpx = dfp.drop(['salary'], axis=1)
with open('noRookies.json', 'wt') as F:
  dfpx.to_json(F, indent=5, orient='records')
with open('noRookiesSalaries.json', 'wt') as F:
  dfp[['PLAYER', 'TEAM', 'salary']].to_json(F, indent=5, orient='records')
```

```python
pp = sns.pairplot(data=dfp, hue='TEAM', height=5,
                  x_vars=['TEAM', 'AGE', 'GP', 'W', 'L', 'MIN', 'OFFRTG', 'DEFRTG',
       'NETRTG', 'AST%', 'AST/TO', 'AST RATIO', 'ORB%', 'DREB%', 'REB%',
       'TO RATIO', 'EFG%', 'TS%', 'USG%', 'PACE', 'PIE', 'POINTS',
       'FGM', 'FGA', '3PM', '3PA', '3P%', 'FTM', 'FTA', 'FT%', 'OREB', 'DREB',
       'REB', 'AST', 'TOV', 'STL', 'PF', 'FP', 'DO2', 'TD3', '+/-', 'YEARS',
       'ALLSTAR', 'ALLSTARPPG'],
                  y_vars=['salary'])
```
